```python
# makes matplotlib (and pandas) plots show up in the notebook
%matplotlib inline

# the following two lines make all plots show up in formats that are high-resolution
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'svg')

# actually import the Pandas module
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
#Get the csv stuff
data = pd.read_csv("plant_bio_project.csv")
data
```

```python
scatter = plt.scatter(data["Group"], data["Root Biomass"])
plt.title("Group Number vs Root Biomass")
plt.xlabel("Group Number")
plt.ylabel("Root Biomass(g)")
plt.show()
```

```python

scatter = plt.scatter(data["plants per pot"], data["Root Biomass"])
plt.title("Actual Number of Plants vs Root Biomasss")
plt.xlabel("Actual Number of Plants")
plt.ylabel("Root Biomass")
plt.xticks(np.arange(1, 5+1, 1.0))
plt.show()


```

```python
scatter = plt.scatter(data["plants per pot"], data["Shoot Biomass"])
plt.title("Actual Number of Plants vs Shoot Biomasss")
plt.xlabel("Actual Number of Plants")
plt.ylabel("Shoot Biomass(g)")
plt.xticks(np.arange(1, 5+1, 1.0))
plt.show()
```

```python
scatter1 = plt.scatter(data["Number"], data["Root Biomass"])
plt.title("Expected Number of Plants vs Root Biomass")
plt.xlabel("Expected Number of Plants")
plt.ylabel("Root Biomass(g)")
plt.show()
```

```python
plt.scatter(data["Root Biomass"], data["Shoot Biomass"])
plt.title("Root Biomass vs Shoot Biomass")
plt.xlabel("Root Biomass(g)")
plt.ylabel("Shoot Biomass(g)")
plt.show()
```

```python
root_biomass_std = np.std(data["Root Biomass"])
print("Root Biomass Std Deviation: ", root_biomass_std)
shoot_biomass_std = np.std(data["Shoot Biomass"])
print("\nShoot Biomass Std Deviation: ", shoot_biomass_std)
#mask = data["plants per pot"] == 1
#print(data[mask])
# print(data[mask])
root_ls = []
root_avg = []
shoot_ls = []
shoot_avg = []
diam_ls =  []
diam_avg = []
for i in range(1,5+1):
    
    mask = data["plants per pot"] == i
    #print(data[mask])
    root_ls.append(np.std(data[mask]["Root Biomass"]))
    root_avg.append(np.mean(data[mask]["Root Biomass"]))
    shoot_ls.append(np.std(data[mask]["Shoot Biomass"]))
    shoot_avg.append(np.mean(data[mask]["Shoot Biomass"]))
    diam_ls.append(np.std(data[mask]["Stem Diameter (mm) "]))
    diam_avg.append(np.mean(data[mask]["Stem Diameter (mm) "]))
    
    
print("\nMean of Root Biomass Per Plants Per Pot: \n", root_avg)
print("\nStd Deviation of Root Biomass Per Plants per pot: \n",root_ls)
print("\nStd Error of Root Biomass", stats.sem(root_ls))
print("\nMean of Shoot Biomass Per Plants Per Pot: \n", shoot_avg)
print("\nStd Deviation of Shoot Biomass Per Plants per pot: \n", shoot_ls)
print("\nStd Error of Shoot Biomass", stats.sem(shoot_ls))
print("\nMean of Root Biomass Per Plants Per Pot: \n", diam_avg)
print("\nStd Deviation of Stem Diameter Per Plants Per Pot: \n", diam_ls)
print("\nStd Error of Stem Diameter", stats.sem(diam_ls))
```

```python
ls = np.arange(1,6)
plt.scatter(ls,root_avg, label="Root Biomass", color = "Green")
plt.plot(ls,root_avg, color = "Green")
plt.scatter(ls, shoot_avg, label = "Shoot Biomass",color = "Red")
plt.plot(ls, shoot_avg, color="Red")
plt.title("Avg Biomass Per Plant Vs Number of Plants Per Pot")
plt.xticks(np.arange(1, 5+1, 1.0))
plt.xlabel("Number of Plants Per Pot")
plt.ylabel("Biomass(g)")
plt.legend()
plt.show()

plt.scatter(ls,diam_avg, color="brown")
plt.plot(ls,diam_avg, color="brown")
plt.xticks(np.arange(1,5+1,1))
plt.xlabel("Number of Plants Per Pot")
plt.ylabel("Stem Diameter(mm)")
plt.title("Stem Diameter vs Number of Plants Per Pot")
plt.savefig("StemDiamvsPPP.png")
plt.show()

```
